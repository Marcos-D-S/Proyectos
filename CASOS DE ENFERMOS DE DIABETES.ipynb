```python
def calculos(x):
  cal1 = kurtosis(x)
  cal2 = skew(x)

  if cal1 > 0.1:
    print(f'El valor de la Kurtosis es: {cal1} y es: leptocurtica')
  elif cal1>= -0.1 and cal1<=0.1:
    print(f'El valor de la Kurtosis es: {cal1} y es: mesocurtica')
  else:
    print(f'El valor de la Kurtosis es: {cal1} y es: platicurtica')

  if cal2 > 0.1:
    print(f'El valor de la asimetria es: {cal2} y es: positiva')
  elif cal2>= -0.1 and cal2<=0.1:
    print(f'El valor de la asimetria es: {cal2} y es: cero')
  else:
    print(f'El valor de la asimetria es: {cal2} y es: negativa')
```

```python
# Calculo de Curtosis y Asimetria de Edad
calculos(Data_Final.EDAD)
```

* HDL: es una sigla que se refiere a "lipoproteína de alta densidad" (del inglés "High-Density Lipoprotein"). El HDL es un tipo de colesterol considerado "bueno" porque ayuda a transportar el colesterol desde otras partes del cuerpo de vuelta al hígado, donde se puede procesar y eliminar del cuerpo.

* LDL: se refiere a "lipoproteína de baja densidad" (del inglés "Low-Density Lipoprotein"). El LDL es un tipo de colesterol considerado "malo" porque puede acumularse en las paredes de las arterias, formando placas que pueden obstruir el flujo sanguíneo y aumentar el riesgo de enfermedades cardiovasculares, como la aterosclerosis, los ataques cardíacos y los accidentes cerebrovasculares.

* PAS se refiere a "Presión Arterial Sistólica". Es uno de los dos valores que se miden para evaluar la presión arterial, siendo el otro la presión arterial diastólica (PAD). La presión arterial se expresa en milímetros de mercurio (mm Hg) y se representa con dos números, como 120/80 mm Hg, donde 120 es la presión sistólica y 80 es la presión diastólica.
* La PAS (SISTOLICA) es un indicador crucial de la salud cardiovascular. Niveles altos de presión sistólica pueden ser un signo de hipertensión, lo que aumenta el riesgo de enfermedades cardíacas, accidentes cerebrovasculares y otros problemas de salud.

Categorías de la Presión Arterial:

* Normal: Menos de 120 mm Hg (PAS) y menos de 80 mm Hg (PAD)
* Elevada: 120-129 mm Hg (PAS) y menos de 80 mm Hg (PAD)
* Hipertensión Fase 1: 130-139 mm Hg (PAS) o 80-89 mm Hg (PAD)
* Hipertensión Fase 2: 140 mm Hg o más (PAS) o 90 mm Hg o más (PAD)
* Crisis Hipertensiva: Más de 180 mm Hg (PAS) y/o más de 120 mm Hg (PAD)

```python
# Creamos una funcion que nos haga los graficos con el enfoque de six sigma para analizar
# valores atipicos

def six_sigma(Dato):
  x = Dato
  x_promedio = Dato.describe()['mean']
  x_desviacion_estandar = Dato.describe()['std']
  plt.figure(figsize=(10,10))
  sns.histplot(data=x)
  plt.axvline(x=x_promedio-x_desviacion_estandar, color='red', linestyle='--')
  plt.axvline(x=x_promedio+x_desviacion_estandar, color='red', linestyle='--')
  plt.axvline(x=x_promedio-6*x_desviacion_estandar, color='magenta', linestyle='--')
  plt.axvline(x=x_promedio+6*x_desviacion_estandar, color='magenta', linestyle='--')
  plt.show()
```

```python
# UTILIZAREMOS EL MAHALANOBIS PARA VER ATIPICOS

def mahalanobis(m=None, data= None, cov=None):
  x_mu = m - np.mean(data)
  if not cov:
    cov = np.cov(data.values.T)
  inv_covmat = np.linalg.inv(cov)
  left = np.dot(x_mu, inv_covmat)
  mahal = np.dot(left, x_mu.T)
  return mahal.diagonal()
```
